```python
import random
import pandas as pd

DATA_PATH = './data/'
DB_PATH = './db/'

df_rest0 = pd.read_json(DATA_PATH + "yelp_restaurants.json")
df_rest0
```

## Preprocessing

```python
df_rest0["city"] = df_rest0["city"].apply(lambda x : x.lower().replace(" ",""))
df_rest0["categories"] = df_rest0["categories"].apply(lambda x : [y.replace(" ","").lower() for y in x.split(',')])
df_rest0.head()
```

```python
df_rest0[df_rest0['attributes'].isnull()]
```

## Rest

```python
df_rest = df_rest0.copy()
df_rest = df_rest[["business_id", "name", "stars", "review_count"]]
df_rest
```

```python
def description_data(df):
    print(df.head())
    print("------- INFO -------")
    print(df.info())
    print("------- DESCRIBE -------")
    print(df.describe())
    print("------- Valeurs NULL -------")
    print(df.isnull().sum())
    print("------- Valeurs manquantes -------")
    print(df.isna().sum())
    print("------- Taille des données -------")
    print(df.shape)
    
description_data(df_rest)
```

```python
count1=0
count2=0
count3=0
count4=0
countNone=0
countnull=0
for index, item in df_rest0.iterrows():
    try:
        if item['attributes'].get('RestaurantsPriceRange2')==None:
            countNone+=1
        if item['attributes'].get('RestaurantsPriceRange2')=='1':
            count1+=1
        if item['attributes'].get('RestaurantsPriceRange2')=='2':
            count2+=1
        if item['attributes'].get('RestaurantsPriceRange2')=='3':
            count3+=1
        if item['attributes'].get('RestaurantsPriceRange2')=='4':
            count4+=1
    except:
        countnull+=1
    #price_range.append(item['attributes']['RestaurantsPriceRange2'])
    
print("None : {}\n null : {}\n 1 : {}\n 2 : {}\n 3 : {}\n 4 : {}".format(countNone, countnull, count1, count2, count3, count4))
```

```python
price_range = []

for index, item in df_rest0.iterrows():
    try:
        p = item['attributes'].get('RestaurantsPriceRange2')
        if p == None:
            price_range.append(None)
        else:
            price_range.append(int(p))
    except:
        price_range.append(None)
        
df_rest['price_range'] = price_range  
df_rest.index.name = "restaurant_id"
df_rest
```

```python
df_rest.to_csv(DB_PATH+'restaurants.csv')
```

## City

```python
cities = []
for city in df_rest0['city']:
    city = city.lower()
    city = city.replace(" ","")
    if city not in cities:
        cities.append(city)

df_city = pd.DataFrame(cities, columns=['city'])
df_city.index.name = "city_id"
df_city
```

```python
df_city.to_csv(DB_PATH+'cities.csv')
```

## Ambiences

```python
list_amb = []
for attr in df_rest0['attributes']:
    try:
        ambs = attr["Ambience"]
        ambs = ambs.replace("{","")
        ambs = ambs.replace("}","")
        ambs = ambs.replace("u'","")
        ambs = ambs.replace("'","")
        ambs = ambs.replace(" ","")
        ambs = ambs.split(",")
        for amb in ambs:
            amb, val = amb.split(":")
            if amb not in list_amb:
                list_amb.append(amb)
    except:
        pass
        
df_ambience = pd.DataFrame(list_amb, columns=['ambience'])
df_ambience.index.name = "ambience_id"
df_ambience
```

```python
df_ambience.to_csv(DB_PATH+'ambiences.csv')
```

## Category

```python
list_cat = []
for cats in df_rest0['categories']:
    for cat in cats:
        cat = cat.replace(" ","")
        cat = cat.lower()
        if cat not in list_cat:
            list_cat.append(cat)

df_category = pd.DataFrame(list_cat, columns=['category'])
df_category.index.name = "category_id"
df_category
```

```python
df_category.to_csv(DB_PATH+'categories.csv')
```

## Located

```python
df_located = df_city.reset_index().merge(df_rest0, how="right", on='city')[["business_id","city_id"]]
df_located = df_located.rename(columns={"business_id": "restaurant_id"})
df_located
```

```python
df_located.to_csv(DB_PATH+'located_relationships.csv')
```

## RestCat

```python
list_cat = []
list_rest = []

 
for val in df_rest0[['business_id','categories']].values:
    for cat in val[1]:
        cat = df_category.index[df_category['category']==cat].to_list()[0]
        list_cat.append(cat)
        list_rest.append(val[0])

df_restCat = pd.DataFrame({"restaurant_id": list_rest, "category_id":list_cat})
df_restCat
```
